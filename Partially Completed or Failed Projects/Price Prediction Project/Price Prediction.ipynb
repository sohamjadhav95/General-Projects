```python
import yfinance as yf

sp500 = yf.Ticker("^GSPC")
sp500 = sp500.history(period = "max")
```

```python
import seaborn as sns
import pandas as pd
from matplotlib import pyplot as plt

sp500
```

```python
sp500.drop('Dividends', axis = 1, inplace=True)
sp500.drop('Volume', axis = 1, inplace=True)
sp500.drop('Stock Splits', axis = 1, inplace=True)
```

```python
sp500.head()
```

```python
sp500["Target (Tomorrow)"] = sp500["Close"].shift (-1)
sp500
```

```python
sp500 = sp500.loc["2020-09-16" :].copy()
sp500.head(3)
sp500.tail(3)
```

```python
sp500.isnull().sum()
```

```python
# Dropping last row

sp500 = sp500[:-1]
sp500.tail()
```

Building Machine Learning Liner Regrassion model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

x = sp500[["Close"]]
y = sp500["Target (Tomorrow)"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size =0.2, random_state = 42)

lr.fit(x_train, y_train)
```

```python
print(lr.score(x_train, y_train)*100)
lr.score(x_test, y_test)*100
```

```python
sns.lineplot(x = 'Date', y = 'Close', data = sp500.tail(50), color = 'Black')
sns.lineplot(x = 'Date', y = 'Target (Tomorrow)', data = sp500.tail(50), color = 'Red')
```
